```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import time

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
from sklearn import svm
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn import manifold
from sklearn.metrics import pairwise_distances
from sklearn.model_selection import cross_validate
```

# Loading data

```python
# Loading dataset 
df = pd.read_csv('dermatology.txt',header=None)

# Rename last column to 'class' 
df.rename(columns={34:'class'},inplace=True)

# Print the 34 features
df.iloc[:,:34].head()
```

# Standardization

```python
# Reenplace lost data with mean of column 
df[33]=pd.to_numeric(df[33],errors='coerce')
df[33]=df[33].fillna(round(df[33].mean(),0))

# Define variables X and y
X = df.iloc[:,:34].values
y = df['class'].values

# Standardization
X = StandardScaler().fit_transform(X)
```

```python
df_X=pd.DataFrame(X)
round(df_X.head(),2)
```

# PCA

## PCA with 0.95 of variance

```python
# pca with minimun 0.95 of variance
pca = PCA(n_components=0.95)
PC = pca.fit_transform(X)

# Cumulative variance
print(pca.explained_variance_ratio_.cumsum())

# Visualization
plt.figure(figsize=(6,4))
a=np.arange(1,len(pca.explained_variance_ratio_.cumsum())+1)
b=pca.explained_variance_ratio_.cumsum()

plt.plot(a,b,marker='o',color='k')

plt.xlabel('Pricipal components')
plt.ylabel('Variance retained')
plt.show()
```

```python
df_pca = pd.DataFrame(PC)
round(df_pca.head(),2)
```

## PCA Rebuild dataset of 2, 4 and 8 components

### n_components = 2

```python
# PCA for 2 components
pca = PCA(n_components=2)
PC2 = pca.fit_transform(X)

# Try to obtain original data
X_inv2 = pca.inverse_transform(PC2)

# Error
print('Variance: {:.3}'.format(pca.explained_variance_ratio_.cumsum()[1]))
print('Loss of 2 PC: {:.3}'.format(np.mean((X-X_inv2)**2)))
print('Sum: {:.3}'.format(pca.explained_variance_ratio_.cumsum()[1]+np.mean((X-X_inv2)**2)))
```

### n_components = 4

```python
# PCA for 4 components
pca = PCA(n_components=4)
PC4 = pca.fit_transform(X)

# Try to obtain original data
X_inv4 = pca.inverse_transform(PC4)

# Error
print('Variance: {:.3}'.format(pca.explained_variance_ratio_.cumsum()[3]))
print('Loss of 4 PC: {:.3}'.format(np.mean((X-X_inv4)**2)))
print('Sum: {:.3}'.format(pca.explained_variance_ratio_.cumsum()[3]+np.mean((X-X_inv4)**2)))
```

### n_components = 8

```python
# PCA for 4 components
pca = PCA(n_components=8)
PC8 = pca.fit_transform(X)

# Try to obtain original data
X_inv8 = pca.inverse_transform(PC8)

# Error
print('Variance: {:.3}'.format(pca.explained_variance_ratio_.cumsum()[7]))
print('Loss of 8 PC: {:.3}'.format(np.mean((X-X_inv8)**2)))
print('Sum: {:.3}'.format(pca.explained_variance_ratio_.cumsum()[7]+np.mean((X-X_inv8)**2)))
```

```python
fig, axes = plt.subplots(1, 3,figsize=(15,4))

ax0, ax1, ax2 = axes.flatten()

ax0.scatter(X_inv2[:,0],X_inv2[:,1])
ax0.scatter(X[:,0],X[:,1])
ax0.set_title('n_components=2')

ax1.scatter(X_inv4[:,0],X_inv4[:,1])
ax1.scatter(X[:,0],X[:,1])
ax1.set_title('n_components=4')

ax2.scatter(X_inv8[:,0],X_inv8[:,1])
ax2.scatter(X[:,0],X[:,1])
ax2.set_title('n_components=8')


plt.show()
```

# Cross-validation and KNeighbors

```python
# Original data
knn_cv=KNeighborsClassifier(n_neighbors=3)
cv_scores=cross_val_score(knn_cv,X,y,cv=8)
print('Original data average accuracy: {:.3}'.format(np.mean(cv_scores)))

# PC8 data
knn_cv=KNeighborsClassifier(n_neighbors=3)
cv_scores8=cross_val_score(knn_cv,PC8,y,cv=8)
print('8 Principal components average accuracy: {:.3}'.format(np.mean(cv_scores8)))
```

```python
a = []
for i in range(1,10):
    knn_cv=KNeighborsClassifier(n_neighbors=i)
    cv_scores=cross_val_score(knn_cv,X,y,cv=8)
    mean=np.mean(cv_scores)
    a.append(mean)
    
plt.plot(range(1,10),a,'b') 
plt.ylabel('Accuracy ')
plt.xlabel('n_neighbors')
plt.title('8 folds')
plt.show()
```

```python
a = []
for i in range(2,10):
    knn_cv=KNeighborsClassifier(n_neighbors=3)
    cv_scores=cross_val_score(knn_cv,X,y,cv=i)
    mean=np.mean(cv_scores)
    a.append(mean)
    
plt.plot(range(2,10),a,'b',color='r') 
plt.ylabel('Accuracy ')
plt.xlabel('Number of folds(KFold)')
plt.title('n_neihbors=3')
plt.show()
```

### Comparation with train_test_split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=4)
print('Train set:',X_train.shape[0])
print('Test set:',X_test.shape[0])

#KNN
neigh=KNeighborsClassifier(n_neighbors = 3).fit(X_train,y_train)


from sklearn import metrics
print("Train set accuracy: ",round(metrics.accuracy_score(y_train,neigh.predict(X_train)),3))
print("Test set accuracy: ", round(metrics.accuracy_score(y_test,neigh.predict(X_test)),3))
```

# Multidimensional scaling

```python
# MDS
seed=np.random.RandomState(seed=3)
similarities=pairwise_distances(X)
mds = manifold.MDS(n_components=2, max_iter=3000, eps=1e-9, 
                   random_state=seed, dissimilarity="precomputed", n_jobs=1)
```

# Clasification Analysis

## KNN 

```python
# n_neighbors=3
knn_cv=KNeighborsClassifier(n_neighbors=3)
cv_3=cross_validate(knn_cv,X,y,cv=8)
print('n_neighbors=3 average accuracy: {:.3}'.format(cv_3['test_score'].mean()))

# n_neighbors=4
knn_cv=KNeighborsClassifier(n_neighbors=4)
cv_4=cross_validate(knn_cv,X,y,cv=8)
print('n_neighbors=4 average accuracy: {:.3}'.format(cv_4['test_score'].mean()))

# n_neighbors=5
knn_cv=KNeighborsClassifier(n_neighbors=5)
cv_5=cross_validate(knn_cv,X,y,cv=8)
print('n_neighbors=5 average accuracy: {:.3}'.format(cv_5['test_score'].mean()))
```

```python
# Training time
print('n_neighbors=3 training time: {:.3}'.format(cv_3['fit_time'].mean()))
print('n_neighbors=4 training time: {:.3}'.format(cv_4['fit_time'].mean()))
print('n_neighbors=5 training time: {:.3}'.format(cv_5['fit_time'].mean()))
print("")
# Prediction time
print('n_neighbors=3 prediction time: {:.3}'.format(cv_3['score_time'].mean()))
print('n_neighbors=4 prediction time: {:.3}'.format(cv_4['score_time'].mean()))
print('n_neighbors=5 prediction time: {:.3}'.format(cv_5['score_time'].mean()))
```

## SVM

```python
svm=svm.SVC(kernel='linear',C=0.025)
cv_smv=cross_validate(svm,X,y,cv=8)

print('SVM average accuracy: {:.3}'.format(cv_smv['test_score'].mean()))
print('SVM training time: {:.3}'.format(cv_smv['fit_time'].mean()))
print('SVM prediction time: {:.3}'.format(cv_smv['score_time'].mean()))
```

## Decision Tree

```python
tree=DecisionTreeClassifier(max_depth = 5)
cv_tree=cross_validate(tree,X,y,cv=8)

print('DecisionTree average accuracy: {:.3}'.format(cv_tree['test_score'].mean()))
print('DecisionTree training time: {:.3}'.format(cv_tree['fit_time'].mean()))
print('DecisionTree prediction time: {:.3}'.format(cv_tree['score_time'].mean()))
```

## AdaBoost

```python
ada = AdaBoostClassifier()
cv_ada=cross_validate(ada,X,y,cv=8)

print('AdaBoost average accuracy: {:.3}'.format(cv_ada['test_score'].mean()))
print('AdaBoost training time: {:.3}'.format(cv_ada['fit_time'].mean()))
print('AdaBoost prediction time: {:.3}'.format(cv_ada['score_time'].mean()))
```

## Gaussian Naive Bayes

```python
gaussian = GaussianNB()
cv_gaussian=cross_validate(gaussian,X,y,cv=8)

print('Gaussian Naive Bayes average accuracy: {:.3}'.format(cv_gaussian['test_score'].mean()))
print('Gaussian Naive Bayes training time: {:.3}'.format(cv_gaussian['fit_time'].mean()))
print('Gaussian Naive Bayes prediction time: {:.3}'.format(cv_gaussian['score_time'].mean()))
```
